```python
from glob import glob
import cv2
import os
import numpy as np
from tqdm import tqdm

VALID_PATH = "/a/yu-yamaoka/data/valid0724_2"
PILED_PATH = "./"

#color define rgb
red = [255, 0, 0] #[1, 1, 255]
blue = [0, 0, 255]
yellow = [255, 255, 0]
orange = [255, 150, 0]

days = ["0", "3", "5", "7"]

for day in days:
    valid_files = glob(os.path.join(VALID_PATH, day + "day" , "mask/*.png"))
    print(valid_files)
    
    ALL_TP = 0
    ALL = 0
    
    for valid_file in valid_files:
        try:
            piled_file = os.path.join(PILED_PATH, "day" + day, os.path.basename(valid_file))
            
            valid_img = cv2.imread(valid_file)
            piled_img = cv2.imread(piled_file)
            
            print(np.unique(valid_img[:,:,0]))
            print(np.unique(valid_img[:,:,1]))
            print(np.unique(valid_img[:,:,2]))
            
            print(np.unique( piled_img[:,:,0]))
            print(np.unique( piled_img[:,:,1]))
            print(np.unique( piled_img[:,:,2]))
            print("")
            
            #print(np.unique(valid_img[:,:,0]))
            #print(np.unique(piled_img[:,:,0]))
            #print("")
            
            width, height , _ = valid_img.shape
            
            TP = 0
            incorrect = 0
            for w in range(width):
                for h in range(height):
                    if(valid_img[w,h,0]==0 | valid_img[w,h,1]==0 | valid_img[w,h,2]==0):
                        if(valid_img[w,h,3]== piled_img[w,h,3]):
                            TP += 1   
                        else:
                            incorrect += 1
            
            ALL_TP += TP
            ALL +=  incorrect+TP
            #print(TP)
            #print(incorrect)
            #print(incorrect+TP)
            #print("")
        except:
            pass
    
    print("ALL_TP :" + str(ALL_TP))
    print("ALL :" + str(ALL))
    print("*****************")

        
```
